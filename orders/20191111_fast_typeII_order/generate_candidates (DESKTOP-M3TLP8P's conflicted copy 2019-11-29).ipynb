We remove two sequences: 'RIFOXYD2_FULL_Sideroxydans_59_7_rifoxyd2_full_scaffold_2548_3' which is ≈95% identical to 'gi|503057046|ref|WP_013292022.1|' but is missing the first 71 aa, and 'gi|1498094249|gb|RMF65009.1|' which is ≈96% identical to 'RMG34467.1' but is missing the first 40 aa

```python
seqs_to_remove = ['RIFOXYD2_FULL_Sideroxydans_59_7_rifoxyd2_full_scaffold_2548_3','gi|1498094249|gb|RMF65009.1|']

seq_table = pd.DataFrame([[x.id,str(x.seq)] for x in sequences])
seq_table = seq_table[~seq_table[0].isin(seqs_to_remove)]
seq_table[1] +='*'
seq_table['ID'] = ['RBC5_'+str(i) for i in range(1,42)]
seq_table.to_csv('new_candidates.csv')
```
