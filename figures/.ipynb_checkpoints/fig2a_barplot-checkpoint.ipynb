```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 22}

plt.rc('font', **font)
```

```python
data = pd.read_csv('../output/01_70p_tree/uclust_all_0.7.csv')
type_data = pd.read_csv('../output/01_70p_tree/rubisco_types_70p.csv')
kin_data = pd.read_csv('../output/00_100p_tree/uclust_all_1_kinetic_data.csv')
data['ID'] = data.Target.apply(lambda x: x.split(' ')[0])
mdata = data.merge(type_data,left_on='ID',right_on='ID')
mdata_kin = mdata.merge(kin_data,left_on='Query',right_on='kinetic_ID')
mdata_kin.loc[mdata_kin.type.isin(['IIIb','unknown']),'type'] = 'III'
measured_types = mdata_kin.groupby('type')['Query'].count()

#measured_types.loc['IIIa'] = 0
#measured_types.loc['IIIc'] = 0
#measured_types.loc['III-like'] = 0

colors = ['#28b463','#e74c3c','#af7ac5','#3498db','#aed6f1','#5d6d7e','#ffbb07']
ax = measured_types.plot.bar(color=colors,logy=True)
ax.set_ylabel('number of measured variants')
ax.set_yticks([0,10,100,200])
ax.set_ylim([0.9,1000])
ax.set_yticks([1,10,100,1000])
ax.set_yticklabels([1,10,100,1000])
plt.savefig('fig2a_inset_num_measured_20190530b.svg')

```

```python
syn_data = pd.read_csv('../output/00_100p_tree/synth_data.csv',header=None)
uclust = pd.read_csv('../output/02_90p_autotrophic_rubisco_tree/auto_uclust_all_0.9_no_outliers_with_type.csv')
len(syn_data)
```

```python
d =uclust.merge(syn_data,left_on='Query',right_on=0,how='left')
d.loc[d.Target_90.isin(d[~d[0].isna()].Target_90.unique()),'flag']=1
```

```python
from Bio import SeqIO, AlignIO, Seq
un_measured = d[(d.flag.isna())&(d.type.isin(['II','II/III']))].Query

#uclust.groupby('type').Target_90.nunique()
x = [record.description for record in SeqIO.parse('../output/00_100p_tree/merged_data_clean.faa', "fasta")]


sequences = []

for record in SeqIO.parse('../output/00_100p_tree/merged_data_clean.faa', "fasta"):
    #print(record.id)
    if record.description in un_measured.values:
        
        sequences.append(record)
with open('../output/00_100p_tree/uclust_all_1.faa', "w") as output_handle:
    SeqIO.write(sequences, 'unmeasured_II_II-III.faa', "fasta")
```
